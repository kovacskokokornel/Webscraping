# Python data structures

## List

They are what they sound like.

```
pi = 3.14
my_list = [1, 2, 'evk', ['szakkollegium', 4], 'spam', pi]
print(my_list)
```

This is how you can append a new element to a list.

```
my_list.append('new item')
print(my_list)
```

.. and this is how to remove elements.

```
my_list.remove('spam')
print(my_list)
```

Lists are ordered, and you can access an item using `[]`.

```
print(my_list[2])
```

Slicing also works here, just as with strings. What strings and lists have in common is that both of them are **iterables** and consist of multiple elements.

```
print(my_list[:3])
```

## Dictionary

Dictionaries are key-value pairs. Keys must be unique across the dictionary. It is a great form of expressing hierarchical relationships.

```
my_dict = {'key1': 'spam', 
           2: 2, 
           'key3': ['s', 'p', 'a', 'm', 3]
}
print(my_dict)
```

Accessing an element by the key.

```
my_dict["key3"]
```

**Use case:** Creating a list of dictionaries. Each dictionary stores details on one person. At the end of the day, this is a database for storing data on students.

```
person_1 = {
    "name": "Joe",
    "age": 22,
    "hair_colour": "blue",
    "course": "webscraping"
}
```

```
person_2 = {
    "name": "Margareth",
    "age": 25,
    "hair_colour": "brown",
    "course": "business consulting"
}
```

```
people = [person_1, person_2]
print(people)
```

## Pandas DataFrame

Think of a dataframe as an excel table. It is a combination of rows and columns.
- It is incredibly easy to filter, search and do any operations on the dataset such as creating a new column based on existing ones. 
- You can import and export to a variaty of formats including CSV or EXCEL making it easy to collaborate with non-pythonic people.
- This is a slippery slope, I just want you to know about the concept. You are not required to learn any operations.

```
import pandas as pd
```

```
df = pd.DataFrame(people)
df
```
