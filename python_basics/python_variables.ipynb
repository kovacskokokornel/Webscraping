## String indexing and slicing

Strings can be indexed, meaning the third index will return the 4th character character in the string. Python starts counting from 0.

````
example = "This is cool!"
print(example[3])
````

Slicing is used to retrieve a part of a string variable. Say you are only interested in the first 3 letters, what do you do? Slice! 

`:` means sort of everything, therefore, `[:3]` means: give me everything until the third character.

````
print(example[:3])
````

`-` meas from the back, so `[:-3]` means: give me everything until the third last character

````
print(example[:-3])
````

You can put the `:` after a number as well, or between numbers

````
print(example[3:])
````

````
print(example[3:9])
````
