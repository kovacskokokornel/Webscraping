# Python control structures

## Conditions

Conditional operators are used to branch the code. It consists of a condition (boolean expression) which can be evaluated as *True* or *False*. Depending on the value, different parts of the code will be executed. The basic structure looks like this:

```
if condition_1:
    statement_1
elif condition_2:
    statement_2
else:
    statement_3
```

````
number = 0.5
print(number < 0)
````

## For loop

A for loop is used to iterate over a bunch of values that are tied together (e.g. string, list). General structure:

```
for variable in iterable:
    statements_1
    statements_2
    ...
```



Iterating over a list of values

````
my_list_of_things = ['a', 'b', 'c', 'd', 
                     1, 2, 3, 4, 
                     'foo', 'bar', 'baz', 'qux']

for item in my_list_of_things:
    print(item)
````

Iterating over string

````
for char in 'this is a python course':
    print(char)
````

Iterating over a range of numbers

````
start = 1
end = 11
step = 1
for number in range(start, end, step):
    print(number)
````

## Let's mix loops with conditions. Print only the even numbers!

````
start = 1
end = 11
step = 1
for number in range(start, end, step):
    if number % 2 ==0:
      print(number)
````
